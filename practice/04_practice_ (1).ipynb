### 1) 성별 데이터 수평 막대 스래프로 시각화

```python
import matplotlib.pyplot as plt
import csv
f1 = open('gender.csv', encoding = 'cp949')
data = csv.reader(f1)

m=[]
f=[]

for row in data:
  if '보람동' in row[0]:
    for i in range(0,101):
      m.append(-int(row[i+3]))
      f.append(int(row[-(i+1)]))
      # -1 값부터 담음 -> 100세 이상 부터 값이 담김


f.reverse()
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.style.use('ggplot') #격자 무늬 스타일로 지정

plt.barh(range(101), m, color= 'steelblue', label='남자')
plt.barh(range(101), f, color= 'tomato', label='여자')
plt.title('세종시 보람동의 남녀 성별 인구 분포')

plt.legend()
plt.show()
```

### 2) 2019년 2월 vs 2020년 2월 연령별 인구 구조 비교

```python
import matplotlib.pyplot as plt
import csv
f2 = open('2008.csv', encoding = 'cp949')
data2 = csv.reader(f2)
f3 = open('2021.csv', encoding = 'cp949')
data3 = csv.reader(f3)
f4 = open('2018.csv', encoding = 'cp949')
data4 = csv.reader(f4)
f5 = open('2020_age.csv', encoding = 'cp949')
data5 = csv.reader(f5)
f6 = open('2019.csv', encoding = 'cp949')
data6 = csv.reader(f6)

result_data2=[]
result_data3=[]
result_data4=[]
result_data5=[]
result_data6=[]



for row in data2:
  if '충청남도 연기군 금남면' in row[0]:
    for i in row[3:]:
      result_data2.append(int(i.replace(',',''))) #인구 수 데이터 정수로 반환 후 result 변수에 담기

for row in data3:
  if '보람동' in row[0]:
    for i in row[3:]:
      result_data3.append(int(i.replace(',',''))) #인구 수 데이터 정수로 반환 후 result 변수에 담기

for row in data4:
  if '보람동' in row[0]:
    for i in row[3:]:
      result_data4.append(int(i.replace(',',''))) #인구 수 데이터 정수로 반환 후 result 변수에 담기

for row in data5:
  if '보람동' in row[0]:
    for i in row[3:]:
      result_data5.append(int(i.replace(',',''))) #인구 수 데이터 정수로 반환 후 result 변수에 담기

for row in data6:
  if '보람동' in row[0]:
    for i in row[3:]:
      result_data6.append(int(i.replace(',',''))) #인구 수 데이터 정수로 반환 후 result 변수에 담기


print(data2)

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.style.use('ggplot') #격자 무늬 스타일로 지정
plt.rc('font', family='NanumBarunGothic') 

plt.plot(result_data2 , color='tomato', label='2008')
plt.plot(result_data4 , color='orange', label='2018')
plt.plot(result_data6 , color='yellow', label='2019')
plt.plot(result_data5 , color='lightgreen', label='2020')
plt.plot(result_data3 , color='purple', label='2021')

plt.title('세종시 보람동의 연도별 인구수 변화 추이')

plt.legend()
plt.show()
```
