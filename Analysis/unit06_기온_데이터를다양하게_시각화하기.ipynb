최고 기온 데이터 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

result=[] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '': #최고 기온 데이터 값이 존재하면 (결측치는 제외)
    result.append(float(row[-1]))

#최고기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(result, bins=100,color='r')
plt.show()
```

10월의 최고 기온 데이터 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

high_aug=[] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  month = row[0].split('-')[1]
  date = row[0].split('-')[2]
  
  if row[-1] != '': #최고 기온 데이터 값이 존재하면 (결측치는 제외)
    if month == '10' and date == '09':
      high_aug.append(float(row[-1]))

#최고기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(high_aug, bins=10,color='orange')
plt.show()
```

1월과 8월의 데이터를 히스토그램으로 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

high_aug=[] #최고 기온 데이터를 저장할 리스트 생성
high_jan=[]
high_apr=[]

for row in data:
  month = row[0].split('-')[1]
  
  if row[-1] != '': #최고 기온 데이터 값이 존재하면 (결측치는 제외)
    if month == '08':
      high_aug.append(float(row[-1]))
    elif month == '01':
      high_jan.append(float(row[-1]))
    elif month == '04':
      high_apr.append(float(row[-1]))

#최고기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(high_aug, bins=100,color='pink',label='august')
plt.hist(high_apr, bins=100,color='lightgreen',label='april')
plt.hist(high_jan, bins=100,color='skyblue',label='january')
plt.legend()

plt.show()
```

1월부터 12월까지 최고 기온 데이터 상자그림으로 표현하기

```python
import matplotlib.pyplot as plt
import random
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

month=[[],[],[],[],[],[],[],[],[],[],[],[]] #월별 데이터 저장할 month 변수 선언

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))
    #월과 같은 번호의 인덱스에 월별 데이터 저장 
    #ex) 1월 ->month[0], 2월 -> month[1]


plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

8월 1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

day=[] #일별 데이터를 저장할 리스트 day변수 선언

for i in range(31):
  day.append([]) #day리스트에 31개 생성

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))
      # 8월 최고 기온 값 저장 day[0] -> 1일

plt.style.use("ggplot")
# 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(10,5),dpi=200)
plt.boxplot(day,showfliers=False) #아웃라이어(이상치) 값 생략
plt.show()
```
