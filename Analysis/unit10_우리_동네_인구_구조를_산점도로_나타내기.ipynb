### 꺾은 선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding="cp949")
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      m.append(int(row[i].replace(',','')))
      f.append(int(row[i+103].replace(',','')))
    break

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.style.use('ggplot') #격자 무늬 스타일로 지정

#꺾은 선 그래프로 남녀 인구수 데이터 표현하기

plt.plot(m, color= 'steelblue', label='남자')
plt.plot(f, color= 'tomato', label='여자')
plt.title(name+'지역의 남녀 성별 인구 분포')

plt.legend()
plt.show()
```

### 막대그래프로 표현하기 <br>
연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding="cp949")
data = csv.reader(f)
result = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.style.use('ggplot') #격자 무늬 스타일로 지정

#꺾은 선 그래프로 남녀 인구수 데이터 표현하기

plt.bar(range(101),result)
plt.title(name+'지역의 남녀 성별 인구 차이')
plt.show()
```

### 랜덤함수를 이용해 서로 다른 100개의 점 산점도 그리기

alpha - 투명도 조절

```python
import matplotlib.pyplot as plt
import random

x=[]
y=[]
size=[]
for i in range(100):
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(50,100))
plt.scatter(x,y,s=size,c=size,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```

### 지역 연령대별 성별 비율을 산점도로 표현하기

```python
import csv
import math
import matplotlib.pyplot as plt

f = open('gender.csv',encoding="cp949")
data = csv.reader(f)

m = []
f = []
size=[] 
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      m.append(int(row[i].replace(',','')))
      f.append(int(row[i+103].replace(',','')))
      size.append(math.sqrt(int(row[i].replace(',',''))+int(row[i+103].replace(',',''))))
      #원의 크기는 인구수의 제곱근 추가

    break

plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.figure(figsize=(10,5),dpi=150)
plt.title(name+'지역의 성별 인구 그래프')
plt.scatter(m,f,s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g') 
#추세선 추가 추세선은 x축기준으로 생성
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
