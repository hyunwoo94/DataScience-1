### 지역명을 입력받아서 연령별 인구수 데이터 시각화하기

```python
import csv
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)

result=[]
name=input('인구 구조가 알고 싶은 지역의 이름(읍/면/동 단위)를 입력해주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',',''))) #인구 수 데이터 정수로 반환 후 result 변수에 담기

#인구 구조 시각화

import matplotlib.pyplot as plt
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic') 

plt.title(name + '지역의 인구 구조')
plt.plot(result)
plt.show()
```

### 성별 데이터 저장하기

성별 데이터 저장하기 - reverse() 사용

```python
import matplotlib.pyplot as plt
import csv
f2 = open('gender.csv', encoding = 'cp949')
data = csv.reader(f2)

m=[]
f=[]

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)]))
      # -1 값부터 담음 -> 100세 이상 부터 값이 담김


f.reverse()
print(m)
print(f)      
```

성별 데이터 저장하기 - for문 2번 사용

```python
import matplotlib.pyplot as plt
import csv
f2 = open('gender.csv', encoding = 'cp949')
data = csv.reader(f2)

m=[]
f=[]

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]:
      m.append(int(i))
    for i in row[106:]:
      f.append(int(i))

print(m)
print(f)
```

### 항아리 그래프를 위해 남성 데이터 음수로 바꾸기

```python
import matplotlib.pyplot as plt
import csv
f2 = open('gender.csv', encoding = 'cp949')
data = csv.reader(f2)

m=[]
f=[]

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101):
      m.append(-int(row[i+3]))
      f.append(int(row[-(i+1)]))
      # -1 값부터 담음 -> 100세 이상 부터 값이 담김


f.reverse()
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.style.use('ggplot') #격자 무늬 스타일로 지정

plt.barh(range(101), m, color= 'steelblue', label='남자')
plt.barh(range(101), f, color= 'tomato', label='여자')
plt.title('신도림지역의 남녀 성별 인구 분포')

plt.legend()
plt.show()
```
