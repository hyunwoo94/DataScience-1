plt.plot(데이터셋) 그래프 그리기

```python
#라이브러리 임포트
import matplotlib.pyplot as plt

#y값이 주어지면 자동으로 x값을 0부터 1씩 증가함.
plt.plot([10,20,30,40,50,60,70,80])

plt.show()
```

plt.plot(x,y) 그래프 그리기

```python
import matplotlib.pyplot as plt

#x값, y값 데이터 입력 
plt.plot([1,2,3,4,5,6],[12,43,25,15,50,20])

plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('plotting')
plt.plot([1,2,3,4,5,6],[12,43,25,15,50,20])

plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('plotting')
plt.plot([10,20,30,40], label='asc') #asc:증가를 의미하는 범례추가
plt.plot([40,30,20,10], label='desc') #desc:감소를 의미하는 범례추가

plt.legend(loc=9)
plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('color')
plt.plot([10,20,30,40], color = 'skyblue',label='skyblue') #하늘색 그래프
plt.plot([40,30,20,10], 'pink',label='pink') #핑크 그래프

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('color')
plt.plot([10,20,30,40], color = 'r',label='red') #기본 색상은 약자로 가능
plt.plot([40,30,20,10], 'k',label='black') #검은색 그래프

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('color')
plt.plot([10,20,30,40], color = 'g',label='green') #초록색 그래프
plt.plot([40,30,20,10], 'y',label='yellow') #노란색 그래프

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('linestyle')
plt.plot([10,20,30,40], color = 'r',linestyle='--',label='dashed') #점선 그래프 - - 
plt.plot([40,30,20,10], 'g', ls = ':',label='dotted') #자잘한 점 그래프 : 

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('marker')
plt.plot([10,20,30,40], 'r.',label='circle') # .:동그라미
plt.plot([40,30,20,10], 'g^',label='triangle up') # ^:삼각형

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#타이틀 속성 추가
plt.title('marker')
plt.plot([10,20,30,40], 'r.-',label='circle') # .:동그라미, -:실선
plt.plot([40,30,20,10], 'g^--',label='triangle up') # ^:삼각형, --:점선

plt.legend()
plt.show()
```

데이터 시각화하기

```python
import csv 
import matplotlib.pyplot as plt
f= open('/content/seoul.csv','r',encoding ='cp949')
data = csv.reader(f)
header = next(data) #헤더 추출
date=[]

result=[]
for row in data:
  if row[-1] != '': #결측치를 제거해서 result 리스트에 넣음
    result.append(float(row[-1]))
    date.append(row[0])
date
plt.figure(dpi=200)
plt.plot(date,result,'r')
plt.show()
```
