데이터 준비하기

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

result=[] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '': #최고 기온 데이터 값이 존재하면 (결측치는 제외)
    result.append(float(row[-1]))

print(result) #최고 기온 리스트 출력
```

```python
print(len(result))
#결측치가 제외 된 데이터 : 39463

```

```python
#8월에 해당되는 값만 추출해서 그래프 시각화
import csv

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

result=[] #최고 기온 데이터중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] != '': #결측치 제거
    if row[0].split('-')[1] == '08': #row[0]:날짜데이터, split으로 나눠서 08월인 것만
      result.append(float(row[-1])) #최고기온 데이터 result에 저장

plt.figure(dpi=200)
plt.plot(result, 'hotpink') #result 리스트의 최고기온 데이터를 딥핑크 그래프로 그리기
```

매년 생일 기준 최고 기온 그래프 그리기

```python
import csv

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

result=[] #최고 기온 데이터중 8월 데이터만 저장할 리스트 생성
date=[]

for row in data:
  if row[-1] != '': #결측치 제거
    if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '09': #매년 10월 9일 데이터 추출
      result.append(float(row[-1])) #최고기온 데이터 result에 저장
      date.append(row[0].split('-')[0])

plt.figure(dpi=200)
plt.plot(date,result, 'lightpink') #result 리스트의 최고기온 데이터를 딥핑크 그래프로 그리기
```

매년 생일 기준 최고 기온, 최저 기온 그래프 그리기

```python
import csv

f = open('seoul.csv',encoding='cp949')

data = csv.reader(f)
next(data) #헤더

high=[]
low=[]
date=[]

for row in data:
  if row[-1] != '' and row[-2] != '':
    if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '09':
      high.append(float(row[-1]))
      low.append(float(row[-2]))
      date.append(row[0].split('-')[0])

plt.figure(figsize=(10,2),dpi=200)
plt.plot(date,high,'hotpink',label='high') #최고기온
plt.plot(date,low,'skyblue',label='low') #최저기온
plt.legend() #범례표시
```

날짜 구간을 좁혀서 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high=[]
low=[]
date=[]

for row in data:
  if row[-1] != '' and row[-2] != '':
    if int(row[0].split('-')[0]) >= 1994:
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '29':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        date.append(row[0].split('-')[0])

#데이터 시각화
#차트 설정
plt.figure(figsize=(15,2),dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
#plot 차트 그리기
plt.plot(date,high,'r',label='high')
plt.plot(date,low,'b',label='low')
plt.title('혁주생일의 기온변화를 그래프로 그리기 ')
plt.legend() #범례 표시
plt.show()
```

최고,평균,최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high=[]
low=[]
avg=[]
date=[]

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '':
    if int(row[0].split('-')[0]) >= 1994:
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '29':
        high.append(float(row[-1]))
        avg.append(float(row[-3]))
        low.append(float(row[-2]))
        date.append(row[0].split('-')[0])

#데이터 시각화
#차트 설정
plt.figure(figsize=(15,2),dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
#plot 차트 그리기
plt.plot(date,high,'r',label='high')
plt.plot(date,avg,'g',label='average')
plt.plot(date,low,'b',label='low')
plt.title('혁주생일의 기온변화를 그래프로 그리기 ')
plt.legend() #범례 표시
plt.show()
```
